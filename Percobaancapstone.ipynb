## Hotel Booking Demand

### Business Problem Understanding

**Context**

Perusahaan yang bergerak di bidang perhotelan ingin melihat adanya pengelolaan reservasi pemesanan kamar. Terdapat berbagai orang yang telah melakukan pemesanan hotel dengan berbagai pilihan tipe kamar. Perusahaan ingin memprediksi adanya pengaruh dari pembatalan dari pemesanan.

Target :

0 : Pembatalan Pemesanan

1 : Tidak Membatalkan Pemesanan

### Attribute Information

| Attribute | Data Type, Length | Description |
| --- | --- | --- |
|Country| Text |Suatu organisasi yang ada di wilayah tertentu baik secara geografis, politik, dan sosial yang berdaulat|
|Market_segment| Text |Kategori pelanggan yang menjadi target atau sumber dari pemesanan kamar hotel.|
|Previous_cancellations	| Long |Jumlah pembatalan sebelumnya oleh pelanggan tertentu |
|Booking_changes| Long |Jumlah dari perubahan dalam permintaan pemesanan oleh pelanggan |
|Deposit_type| Text |Tipe deposit yang dipilih oleh pelanggan ketika melakukan pemesanan hotel|
|Days_in_waiting_list| Long | Jumlah dari hari pemesanan yang berada didalam daftar tunggu sebelum akhirnya dikonfirmasi atau dibatalkan|
|Customer_type| Text |Tipe atau kategori pelanggan yang melakukan pemesanan hotel|
|Reserved_room_type| Text | Jenis atau kategori kamar yang telah dipesan oleh pelanggan untuk periode tertentu|
|Required_car_parking_spaces| Long |Jumlah tempat parkir mobil yang diminta atau dibutuhkan oleh pelanggan saat melakukan pemesanan kamar hotel|
|Total_of_special_requests| Long |Jumlah total permintaan khusus atau permintaan tambahan yang diajukan oleh pelanggan saat melakukan pemesanan kamar hotel|
|Is_canceled| Long |0 – Pembatalan Pemesanan, 1 – Tidak Membatalkan Pemesanan|
